```python
def euclidSim(inA,inB,K):
    d=[]
    for i in range(len(inB)):
        sim= np.array(1.0/(1.0 + la.norm(inA - inB[i])))
        d.append(sim)
        d_arr= np.array(d)
        idx = np.argsort(d_arr)
        reverse_array = idx[::-1]
    return reverse_array[:K],d_arr[reverse_array[:K]]



def pearsonSim(inA,inB,K):
    d=[]
    for i in range(len(inB)):
        if len(inA) < 3 : 
            return 1.0
        else:
            sim= 0.5+0.5*corrcoef(inA, inB[i], rowvar = 1)[0][1]
            d.append(sim)
            d_arr= np.array(d)
            dx = np.argsort(d_arr)
            idx = np.argsort(d_arr)
            reverse_array = idx[::-1]
    return reverse_array[:K],d_arr[reverse_array[:K]]


from scipy.spatial.distance import jaccard
from sklearn.metrics import jaccard_score

def jaccard_sim(inA,inB,K):
    d=[]
    for i in range(len(inB)):
        similarity = jaccard_score(inA, inB[i])
        d.append(similarity)
        d_arr= np.array(d)
        dx = np.argsort(d_arr)
        idx = np.argsort(d_arr)
        reverse_array = idx[::-1]
    return reverse_array[:K],d_arr[reverse_array[:K]]
    



```

```python
from numpy import *
from numpy import linalg as la
import numpy as np

def euclidSim(inA,inB):
    return 1.0 / (1.0 + la.norm(inA - inB))

def pearsonSim(inA,inB):
    if len(inA) < 3 : return 1.0
    return 0.5 + 0.5 * corrcoef(inA, inB, rowvar = 0)[0][1]

def cosineSim(inA,inB):
    num = float(inA.T * inB)
    denom = la.norm(inA)*la.norm(inB)
    return 0.5 + 0.5 * (num / denom)

def jaccardsim(inA,inB):
    from scipy.spatial.distance import jaccard
    from sklearn.metrics import jaccard_score

    return jaccard_score(inA, inB)
    
```

```python
def print_most_similar_users(dataMat, user, k, metric):
    print ("User", user ,":\n")
    knn_users = []
    for all_users in range(len(dataMat)):
        user_sim = np.array((metric(dataMat[user,:],dataMat[all_users,:])))
        knn_users.append(user_sim)
        d_arr= np.array(knn_users)
        idx= np.argsort(d_arr)
        reverse_array = idx[::-1]
        knn_user= np.delete(reverse_array[:k+1],0)
        knn_user_sim = d_arr[knn_user[:k+1]]
    return knn_user, knn_user_sim
```

```python
def print_most_similar_items_based_on_knn_users(dataMat, user, k, metric):
    # for a specific user this gives the simmilarity btween items
    knn_users, knnusers_sim= print_most_similar_users(dataMat,user , k, metric)
    item_user_bough= []
    select_user_purchase = dataMat[user]
    
    knn_matrix=[]
    
    for users_knn in knn_users:
        knn_matrix.append((dataMat[users_knn]))
    
    #This function give what item user bought
    for select_user_purchase in range((dataMat[user].shape[1])):
        if dataMat[user,select_user_purchase] >0:
            item_user_bough.append(select_user_purchase)
    
    # for the items bought find the similarites between the items with KKN_USERS
    for itemes_bought in item_user_bough:
        item= itemes_bought
     
    
        d=[]
        all_knn_matrix= np.concatenate((knn_matrix), axis=0)
        for knn_users_ in range((all_knn_matrix.shape[1])):
            sim = np.array((metric(all_knn_matrix[:,item],all_knn_matrix[:,knn_users_])))
            #print(sim)
            d.append(sim)
            d_arr= np.array(d)
            idx= np.argsort(d_arr)
            reverse_array = idx[::-1]
            knn_item= np.delete(reverse_array[:k+1],0)
            knn_sim = d_arr[knn_item[:k+1]]
            
    return knn_item ,knn_sim,item_user_bough
```

```python
def train_item_item(dataMat,metric ):
    number_of_feature= dataMat.shape[1]
    # Create a matrix with all entries = NDocs
    sim_matrix= []

    for features_ in range((dataMat.shape[1])):
        for rated_joke in range((dataMat.shape[1])):
            sim = (metric(dataMat[:,features_],dataMat[:,rated_joke]))
            sim_matrix.append(sim)
            #newarr = sim_matrix.reshape(100, 100)
    
    return np.array(sim_matrix).reshape(number_of_feature, number_of_feature)
```
